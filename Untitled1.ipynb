```python
import pandas as pd
import os
import geopandas as gpd
```

```python
path = '/home/sgirtsou/Documents/processes/pre-processing/processed'
daily = []
```

```python
for file in os.listdir(path):
    if os.path.isdir(os.path.join(path,file)):
        date = str(file)
        for file1 in os.listdir(os.path.join(path,file)):
            if os.path.isdir(os.path.join(path,file,file1)):
                 shp_path = (os.path.join(path,file,file1,file1+'.shp'))
                 daily.append(shp_path)
```

```python
attica_cells = pd.read_csv('/home/sgirtsou/Documents/vravrona/attica_cells.csv')
```

```python
attica_ids = attica_cells.id.to_list()
```

```python
df1 = gpd.read_file(daily[0])
df2 = df1[['id','Comb_proba']].loc[df1['id'].isin(attica_ids)].reset_index()
del df1
```

```python
df3 = gpd.read_file(daily[1])
df4 = df3[['id','Comb_proba']].loc[df3['id'].isin(attica_ids)].reset_index()
del df3
```

```python
df = df2.append(df4,ignore_index=True)
```

```python
for i,path_row in enumerate(daily):
    print(i)
    if i == 0:
        df0 = gpd.read_file(path_row)
        df1 = df0[['id','Comb_proba']].loc[df0['id'].isin(attica_ids)].reset_index()
        del df0
    else:
        df0 = gpd.read_file(path_row)
        df0_i = df0[['id','Comb_proba']].loc[df0['id'].isin(attica_ids)].reset_index()
        df1 = df1.append(df0_i)
        del df0
        del df0_i
```

```python
df1
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
labels = ['no risk','low risk','inter risk','high risk','very high risk']
```

```python
df1['binned'] = pd.cut(df1['Comb_proba'], bins=bins,labels=labels)
```

```python
df_grouped = df1.groupby(by=['id','binned']).size().reset_index()
```

```python
df_grouped[df_grouped.binned=='very high risk']
```

```python
idx = df_grouped.groupby(['id'])[0].transform(max) == df_grouped[0]
```

```python
ids_risk = df_grouped[idx]
```

```python
ids_risk
```

```python
attica_cells_risk = attica_cells.merge(ids_risk, on = 'id')
```

```python
attica_cells_risk.to_csv('/home/sgirtsou/Documents/vravrona/attica_cells_risk.csv')
```
